# Churn Data Analysis

## Scenario: Telecommunications Churn

In the telecommunications industry, customers can choose from multiple service providers and actively switch from one provider to another. Customer “churn” is defined as the percentage of customers who stopped using a provider’s product or service during a certain time frame. In this highly competitive market, some telecommunications industries can experience average annual churn rates as high as 25 percent. Given that it costs 10 times more to acquire a new customer than to retain an existing one, customer retention has now become even more important than customer acquisition.

For many providers, retaining highly profitable customers is the number one business goal. To reduce customer churn, telecommunications companies need to predict which customers are at high risk of churn.

## Question
> What are the affects of tenure and GB usage on churn rate?

> Stakeholders in the organization would benefit from an analysis of the data because an analysis brings actionable insights
> that the company can make use of to improve business function. For example, they can find what factors greatly correlate and contribute to customer churn rate, and the organization can work on fixing those issues to lower the churn rate. 

> The data needed for this analysis is Bandwidth_GB_Year, Tenure, and Churn.

```python
# Importing the packages that will be used for this analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
# Loading in our dataset in the form of a Pandas DataFrame. We name the variable 'df' for DataFrame.
df = pd.read_csv('Cleaned Churn Data\churn_clean.csv')

# Displaying the first few rows of df to make sure it imported correctly
df.head()
```

```python
# Correlation between all variables
df.corr()
```

```python
# Standard deviation of the dataframe
df.std()
```

```python
# Creating the dataframe for the analysis of the question
df_analysis = df[['Tenure', 'Bandwidth_GB_Year','Churn']].copy()

# Displaying the first few rows of the new dataframe
df_analysis.head()
```

## T-Test

```python
# T-Test for bandwidth usage and churn rate
yes_churn = df_analysis.loc[df_analysis['Churn'] == 'Yes','Bandwidth_GB_Year']
no_churn = df_analysis.loc[df_analysis['Churn'] == 'No','Bandwidth_GB_Year']
print(ttest_ind(yes_churn,no_churn))
```

> This test shows that lower gb usage equals a much higher churn rate than those that have high gb usage.

> The pvalue is 0 which gives us a very high certainty that this result is correct

```python
# T-Test for tenure and churn rate
yes_churn = df_analysis.loc[df_analysis['Churn'] == 'Yes','Tenure']
no_churn = df_analysis.loc[df_analysis['Churn'] == 'No','Tenure']
print(ttest_ind(yes_churn,no_churn))
```

> This test shows that those with a shorter tenure have a much higer churn rate than those with a longer tenure.

> The pvalue is 0 which gives us a very high certainty that this result is correct
